```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
exoplanet_hunting = pd.read_csv('DATA/exoplanet_hunting/exoTrain.csv', low_memory=False)
```

```python
print(exoplanet_hunting.shape)
exoplanet_hunting.head()
```

Each row represents one host star, the columns corresponding to its magnitude at a moment in time.

---

### Evolution of magnitude plotted over time

```python
def scale_row(row, rng=(-1,1)):
    return np.interp(row, (row.min(), row.max()), rng)
```

```python
plt.figure(figsize=(25,10))
plt.plot(scale_row(exoplanet_hunting.iloc[0,1:]))
plt.show()
```

#### The following two graphs demonstrate the target effect

```python
plt.figure(figsize=(25,10))
plt.plot(scale_row(exoplanet_hunting.iloc[1,1:]))
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.plot(scale_row(exoplanet_hunting.iloc[7,1:]))
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.plot(scale_row(exoplanet_hunting.iloc[13,1:]))
plt.show()
```
